# 1. Explorar datasets

Se seleccionó el conjunto de datos [EEG Motor Movement/Imagery Dataset v1.0.0 (physionet.org)](https://physionet.org/content/eegmmidb/1.0.0/) para el presente Challenge. Este dataset, proporcionado por los desarrolladores del sistema de instrumentación BCI2000,se utiliza en investigaciones relacionadas con interfaces cerebro-computadora. Contiene señales EEG de 64 canales provenientes de individuos que llevaron a cabo diversas tareas motoras o ejercicios de imaginación motora [1]. Este dataset incluye más de 1.500 registros de EEG, con duraciones de uno y dos minutos, recolectados de 109 participantes voluntarios.

## 1.1. Protocolo experimental

Los participantes realizaron diversas tareas motoras e imaginación motora mientras se registraban sus señales EEG de 64 canales utilizando el sistema BCI2000 (http://www.bci2000.org). Cada individuo completó 14 sesiones experimentales, que incluyeron dos pruebas de referencia de un minuto (una con los ojos abiertos y otra con los ojos cerrados), así como tres repeticiones de dos minutos para cada una de las siguientes cuatro tareas:

- Tarea 1: Aparece un objetivo a la izquierda o a la derecha de la pantalla. El participante **abre y cierra el puño** correspondiente hasta que el objetivo desaparezca y, posteriormente, **se relaja**.
- Tarea 2: Aparece un objetivo a la izquierda o a la derecha de la pantalla. El participante **imagina abrir y cerrar el puño** correspondiente hasta que el objetivo desaparezca y, después, **se relaja**.
- Tarea 3: Aparece un objetivo en la parte superior o inferior de la pantalla. El participante **abre y cierra ambos puños** (si el objetivo está arriba) **o ambos pies** (si el objetivo está abajo) hasta que el objetivo desaparezca y, luego, **se relaja**.
- Tarea 4: Aparece un objetivo en la parte superior o inferior de la pantalla. El participante **imagina abrir y cerrar ambos puños** (si el objetivo está arriba) **o ambos pies** (si el objetivo está abajo) hasta que el objetivo desaparezca y, después, **se relaja**.

En resumen, las sesiones experimentales fueron las siguientes:

- 1 - Línea de base, ojos abiertos
- 2 - Línea de base, ojos cerrados
- 3-6 - Tareas 1 a 4 (primera vez)
- 7-10 - Tareas 1 a 4 (segunda vez)
- 11-14 - Tareas 1 a 4 (tercera vez)

Los datos se proporcionan en **formato EDF+** (que incluye 64 señales de EEG, cada una con una frecuencia de muestreo de 160 Hz, y un canal de anotaciones). Los archivos .event y los canales de anotaciones de los archivos .edf correspondientes contienen información idéntica.

Cada anotación incluye uno de los tres códigos (T0, T1 o T2):

- T0: reposo
- T1: inicio del movimiento (real o imaginario) del puño izquierdo (en las sesiones 3, 4, 7, 8, 11 y 12) o de ambos puños (en las sesiones 5, 6, 9, 10, 13 y 14)
- T2: inicio del movimiento (real o imaginario) del puño derecho (en las sesiones 3, 4, 7, 8, 11 y 12) o de ambos pies (en las sesiones 5, 6, 9, 10, 13 y 14)

# 2. Seleccionar un dataset y justificar

La imaginación motora (MI) es un proceso cognitivo que implica la simulación mental de acciones motrices sin la ejecución real del movimiento [3]. Se ha demostrado que activa circuitos neuronales similares a los que se activan durante los movimientos reales, lo que constituye una base esencial para el desarrollo de interfaces cerebro-ordenador (BCI) [4, 5]. Las BCI son tecnologías emergentes que permiten la comunicación directa entre el cerebro humano y dispositivos externos, de forma que se traduce la actividad cerebral en órdenes [5]. La aplicación de las MI a los sistemas BCI ofrece grandes posibilidades de mejorar la calidad de vida de las personas con diversas discapacidades motoras y trastornos neurológicos [3].

La principal motivación para el desarrollo de las BCI es la creciente necesidad de métodos alternativos de comunicación y control para las personas que padecen enfermedades debilitantes [3, 6, 7, 8]. Estas afecciones, como las lesiones medulares, los derrames cerebrales o la esclerosis lateral amiotrófica (ELA), suelen provocar graves deficiencias motoras que dejan a los pacientes con una capacidad limitada o nula para comunicarse o interactuar con su entorno [3, 9]. Las BCI ofrecen una solución prometedora para superar las vías neuronales alteradas y devolver cierto grado de autonomía a estas personas, mejorando así su bienestar general y reduciendo la carga de los cuidadores [5, 9].

La electroencefalografía (EEG) es una técnica no invasiva muy utilizada para medir la actividad cerebral y se ha empleado ampliamente en los sistemas BCI basados en MI [3, 5, 6, 7]. Las señales EEG tienen la ventaja de ser relativamente fáciles de adquirir y pueden proporcionar información en tiempo real sobre las intenciones del usuario [10]. Sin embargo, estas señales suelen ser ruidosas y complejas, por lo que es necesario utilizar sofisticadas técnicas de procesamiento de señales y algoritmos de aprendizaje automático para extraer información significativa que permita clasificar con precisión el movimiento visualizado [3, 7, 10].

En los últimos años, los avances en las técnicas de aprendizaje automático y aprendizaje profundo han dado lugar a mejoras sustanciales en el rendimiento y la fiabilidad de los sistemas BCI basados en MI [3, 6, 7]. Estos algoritmos son capaces de aprender patrones complejos a partir de grandes conjuntos de datos, lo que permite una clasificación y predicción más precisas de las intenciones del usuario. Además, los algoritmos de aprendizaje automático pueden adaptarse y ajustarse a las necesidades específicas de cada usuario, lo que permite crear sistemas BCI personalizados que atienden a los requisitos únicos de cada usuario [5].

El objetivo que escogimos para este Challenge es el de **explorar el potencial de la Discrete Wavelet Transform, Complejidad de Hjorth y Operador de Taeger para el procesamiento de las señales electroencefalográficas y el aprendizaje automático supervisado tanto por los métodos convencionales estudiados en clase como por los métodos de ensamble para el desarrollo de un algoritmo para un sistema BCI robusto y eficiente basado en MI para la clasificación de la imaginación de apertura y cierre de la mano izquierda o derecha**.

### 3.2.3. Discrete Wavelet Transform, Hjorth Complexity y Taeger Energy Operator

Para cada conjunto de señales EEG aplicamos la DWT para obtener finalmente las bandas 16-32 Hz y 8-16 Hz, estas corresponden a los ritmos beta y mu de las ondas cerebrales, que como se mencionó anteriormente, son útiles para la imaginación motora. Con la finalidad de obtener mejores resultados, se utilizará los coeficientes de Daubechies ortogonales de orden 16, particularmente usamos esta Wavelet pues permite discernir información que cambia en instantes de tiempos medianamente separados, esto sucede porque la actividad cerebral de la imaginación del movimiento no es instantánea como lo pueden ser los ataques epilépticos, sin embargo también se debe considerar ciertos cambios instantáneos, por ello se usa el de orden 16 [14, 15, 16, 17].

La DWT consiste en un banco de filtros que continuamente realizan un proceso de submuestreo, tal y como se muestra en la Figura 2, donde el filtro está definido por el Wavelet escogido. Nosotros nos quedamos con el tercer nivel de detalle y el cuarto nivel de detalle, pues como se mencionó previamente, permiten extraer las ondas cerebrales más significativas. Por otro lado, previo al filtrado, se realizó un resampleo de la señal para que la nueva frecuencia de muestreo sea 128 Hz, de tal forma que al aplicar los filtros se llega a la banda deseada



Figura 2. Banco de filtros para Discrete Wavelet Transform

Por otro lado, también se usará la complejidad de Hjorth, este es un parámetro para la extracción de características en el tiempo, de esta se obtienen 3 parámetros [18]:
- Actividad: representa la potencia de la señal tomando la variance en función del tiempo.



- Mobilidad: parámetro que representa la frequencie mediana o la proporción de la desviación estándar del espectro de potencia. Este se define como la raiz de la varianza de la primera dervidad de la señal dividido por la varianza de la señal.



- Complejidad: representa la razón de cambio de la frecuencia. Este parámetro compara la similaridad de la señal con una sinusoide, donde el valor converge a 1 si las señales son más similares.



Además de esto, el Operador de la Energía de Taeger es un operador no lineal que refleja el cambio instantáneo de la energía. Para una señal sinusoidal $x(t) = A\cos{wt+\phi}$, TEO denotado como $\Psi{[x(t)]}$, es definido como $\Psi{[x(t)]}=[x'(t)]^2-x(t)x''(t)$. Para este trabajo se toma un equivalente en tiempo discreto [19]. Considerarmos $x'[n]=\frac{x[n]-x[n-1]}{T}$ y que $T ≃ 1$ para obtener la siguiente ecuación de diferencias para TEO: $\Psi{[x(t)]}=[x(n)^2]-x(n-1)x(n+1)$. Para representar el cálculo de TEO con valores únicos, se ha utilizado en el estado del arte (para otra ataques epilépticos) la media y la varianza [20].

```python
import numpy as np
from scipy.signal import welch
from scipy.integrate import simps

def TEO(eeg):
    TEO = np.zeros(len(eeg)-2)
    for j in range(1, len(eeg)-1):
        TEO[j-1] = eeg[j]**2 - eeg[j-1]*eeg[j+1]
    return TEO


def complexity_hjorth(signal):
    dx = np.diff(signal)
    ddx = np.diff(dx)
    x_var = np.var(signal)
    dx_var = np.var(dx)
    ddx_var = np.var(ddx)
    mobility = np.sqrt(dx_var / x_var)
    complexity = np.sqrt(ddx_var / dx_var) / mobility
    return complexity, mobility, x_var
```

```python
import pywt
from scipy.signal import resample
from scipy.stats import kurtosis

s01_dwt = []

for event in filtered_sigs:
    sigs = []
    for sig in event:
        sig = resample(sig, int(len(sig)*0.8))
        coeffs = pywt.wavedec(sig, "db16", level=4)
        cA4, cD4, cD3, cD2, cD1 = coeffs
        f1 = np.mean(np.abs(cD4))
        f2 = np.mean(np.abs(cD3))
        f3 = np.mean(np.square(cD4))
        f4 = np.mean(np.square(cD3))
        f5 = np.std(cD4)
        f6 = np.std(cD3)
        f7 = f1/f2
        f8 = kurtosis(cD4)
        f9 = kurtosis(cD3)
        t = TEO(sig)
        f10 = np.mean(t)
        f11 = np.var(t)
        f12, f13, f14 = complexity_hjorth(sig)
        sigs.append([f1, f2, f3, f4, f5, f6, f7, f8, f9, f10, f11, f12, f13, f14])
    s01_dwt.append(sigs)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

Generamos un Pipeline para realizar el escalado de las variables, luego seleccionamos las características que tengan una varianza explicada mayor a 0.25, luego se obtienen los componentes principales tomando 100 componentes y se realiza un análisis de componentes independientes igualmente tomando 100 componentes.

```python
from sklearn.preprocessing import RobustScaler
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import VarianceThreshold
from sklearn.decomposition import PCA, FastICA
pipe = Pipeline([
    ("scale", RobustScaler()),
    ("var_t", VarianceThreshold(threshold=(0.5*(1-0.5)))),
    ("pca", PCA(n_components=100, random_state=42)),
    ("ica", FastICA(n_components=100, random_state=42))
])
```

Entrenamos el pipelina para obtener las matriz de características final.

```python
X_train = pipe.fit_transform(X_train)
X_test = pipe.transform(X_test)
X_train.shape, X_test.shape
```

A partir de los resultados obtenidos, el modelo supervisado mostró un rendimiento ligeramente superior al mejor modelo de ensamble. Sin embargo, es crucial mencionar que los hiperparámetros de CatBoost no han sido optimizados, lo que significa que aún hay margen para mejorar su desempeño. Es razonable, ya que el modelo CatBoost es más eficiente en la identificación de relaciones complejas. No obstante, esto podría ser contraproducente, ya que puede generar sobreajuste en el conjunto de entrenamiento. Para abordar este problema, vamos a comparar la precisión de ambos modelos utilizando el conjunto de entrenamiento.

```python
print(accuracy_score(y_train, ensemble_model.predict(X_train)))
print(accuracy_score(y_train, supervised_model.predict(X_train)))
```

Observamos que, en el caso de CatBoost, el modelo experimentó sobreajuste, lo que explica los resultados insatisfactorios en el conjunto de prueba, mientras que el modelo de SVM mantiene una precisión consistente. Esto sugiere que, aunque los modelos de ensamble son altamente competentes para descubrir relaciones complejas entre variables, también presentan un riesgo considerable de sobreajuste, especialmente cuando el número de muestras es relativamente bajo. A pesar de contar con 1239 muestras, esta cantidad sigue siendo insuficiente para modelos de Deep Learning, por ejemplo.

Además, los resultados obtenidos no son adecuados para aplicaciones clínicas, ya que la precisión es muy baja, lo que provocaría numerosos errores al intentar implementar estos modelos en entornos reales.

# 7. Conclusiones

- Una limitación principal de este Challenge fue la incapacidad de usar el método `crop_by_annotations()` en varias sesiones, lo que disminuyó considerablemente la cantidad de señales disponibles para entrenar los modelos. Aunque se pudo solucionar empleando ventanas en la señal, no se implementó debido a restricciones de tiempo.
- La frecuencia de muestreo de 160 Hz es bastante baja en comparación con otras aplicaciones de Imaginación Motora, donde normalmente se utilizan frecuencias de muestreo de 250 Hz o más. Una frecuencia de muestreo mayor ayuda a evitar errores de aliasing que pueden afectar la descomposición de las señales en diferentes bandas de frecuencia, ya que estarían contaminadas por información no perteneciente a las ondas cerebrales deseadas.
- Los métodos DWT, complejidad de Hjorth y operador de energía de Taeger no resultaron precisos en la clasificación de imaginación motora en señales EEG muestreadas a 160 Hz, posiblemente debido a la contaminación de la señal por la frecuencia de muestreo empleada pues ambas herramientas demostraron ser útiles en esta misma aplicación o en otras.
- Las bajas precisiones de los modelos también pueden atribuirse a la necesidad de otros métodos de preprocesamiento. Uno de los enfoques más importantes para mejorar la relación señal-ruido en los EEG es aplicar una referencia bipolar, pero en el conjunto de datos, los electrodos A1 y A2 fueron eliminados, lo que dificulta la reducción del ruido. Alternativas como Independent Component Analysis y Common Spatial Pattern podrían ser soluciones viables, pero su complejidad va más allá del alcance del curso.
- La mayoría de los modelos con precisión superior al 70% en este conjunto de datos se lograron mediante modelos de Deep Learning. No se han reportado algoritmos de alta precisión utilizando modelos de Machine Learning a menos que se hayan aplicado técnicas de preprocesamiento sofisticadas, las cuales, como se mencionó anteriormente, exceden el alcance del curso.
- Dada la naturaleza de los resultados, es probable que agregar el conjunto de datos que contiene las etapas de relajación hubiera incrementado la precisión. No obstante, esto podría ser contraproducente para distinguir entre los movimientos imaginarios, ya que la diferencia entre ellos es mínima en comparación con la diferencia respecto a las etapas de relajación. Por lo tanto, es posible que los modelos predigan correctamente la relajación, pero no los movimientos imaginados.
- Los modelos de ensamble como CatBoost, debido a su capacidad para encontrar interacciones complejas entre variables, pueden ser propensos al sobreajuste en comparación con modelos supervisados, como lo fue en este caso el Support Vector Machine. Por ello, se debe aumentar la cantidad de datos (o reducir el conjunto de entrenamiento) para evitar este problema.
- Las precisiones alcanzadas por los modelos obtenidos no son lo suficientemente altas como para ser aplicables en contextos clínicos.
- Este dataset podría ser utilizado en un proyecto del curso debido a la complejidad que implica el manejo de las señales EEG y el preprocesamiento adecuado de las mismas.
